## Bearbeitung der Reviews
Hier werden aus den 4 einzelnen `combined_data.csv` Dateien eine groÃe Datei gemacht, die alles enthÃ¤lt.
Da in den originalen Dateien die Filme zu den Reviews so gekennzeichnet wurden, dass vor allen Reviews zu einem Film in einer Zeile die Movie_ID und dann ein ':' geschrieben wurde.

Das wÃ¤ren Reviews zum Film mit der ID 1.
```
1:
1488844,3,2005-09-06
822109,5,2005-05-13
885013,4,2005-10-19
```

Da Pandas nicht gut damit umgehen kann, erstellen wir eine neue Datei, die als extra Spalte neben jeder Review auch die ID des Films angibt. Diese speichern wir unter `input/combined_data.csv` ab.

````python
def extractData():
    # Erstelle File und schreibe header
    with open(f"combined_data.csv", 'w') as file:
        file.write('Movie_ID,User_ID,Ratings,ReleaseDate\n')

    # Ãffne das selbe File in append mode und fÃ¼ge neue Daten hinzu 
    with open(f"combined_data.csv", 'a') as out_file:
        for i in range(1, 5):
            with open(f"../input/netflix-prize-data/combined_data_{i}.txt", 'r') as in_file:
                lines = in_file.read().split('\n')

            new, movie_id = [], ''
            for line in lines:
                if line == '':
                    continue
                if ':' in line:
                    movie_id = line.replace(':', '')
                    continue
                new.append(f"{movie_id},{line}")

            out_file.write('\n'.join(new) + '\n')
````

## Bearbeitung der Filme
Hier wird die Datei, welche Details zu den Filmen enthÃ¤lt von einem csv manuell in ein tsv umgewandelt, da die Filmnamen ',' enthalten und der pandas Tokenizer das nicht verarbeiten kann. Zudem Ã¤ndern wir das character encoding zu utf-8.

````python
def formatMovieData():
    with open('../input/netflix-prize-data/movie_titles.csv', 'r', encoding='ISO-8859-1') as file:
        lines = file.read().split('\n')

    new = list(map(lambda x: x.replace(',', '\t', 2), lines))

    with open('movie_titles.tsv', 'w', encoding='UTF-8') as file:
        file.write('Movie_ID\tYear\tTitle\n' + '\n'.join(new))
````

````python
df_metadata = pd.read_csv('../input/netflix-shows/netflix_titles.csv', usecols=['title', 'type','release_year', 'rating', 'listed_in', 'description']).dropna()
````

````python
print(df_metadata.shape)
df_metadata.head()
````

````python
#df_metadata = df_metadata[df_metadata['type'] == 'Movie'].drop('type', axis=1)
````

````python
print(df_metadata.shape)
df_metadata['Index'] = range(df_metadata.shape[0])
df_metadata.set_index('Index', inplace=True)
df_metadata.head()
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
#removing stopwords
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
df_metadata['description'] = df_metadata['description'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df_metadata['description'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
````

````python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(df_metadata.index, index=df_metadata['title']).drop_duplicates()
````

````python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df_metadata.iloc[movie_indices]
````
